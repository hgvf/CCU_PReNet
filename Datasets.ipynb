```python
import os
import numpy as np
import random
import torch
import cv2
import torch.utils.data as udata
import torch.utils.data.dataset as Datasets
from PIL import Image
from torchvision import transforms
from IPython.display import display
```

```python
# 自訂 datasets

class PReNet_train_datasets(torch.utils.data.Dataset):
    
    def __init__(self, transform):
               
        # datasets 存放路徑
        self.path = ''
        self.input_path = os.path.join(self.path, 'RAIN')
        self.target_path = os.path.join(self.path, 'GT')
        
        imgs = []

        # 蒐集所有檔名
        files = os.listdir(self.input_path)
        
        for f in files:
            imgs.append((f, int(0)))
        
        self.imgs = imgs
        self.len = len(imgs)
        self.transform = transform
        self.files = files

    def __getitem__(self, index):
        
        path, label = self.imgs[index]
        
        # 取得某照片的檔名 & 路徑，並開啟圖片
        input_path = os.path.join(self.input_path, self.files[index])
        input_image = Image.open(input_path)
        #print('=-=-=-=-=-=-=-=-=-=-')
        #print('index : ', index)
        #print('-=-=-=-=-=-=-=-=-=-=-')
        #display(input_image)
        target_path = os.path.join(self.target_path, self.files[index])
        target_image = Image.open(target_path)
        #display(target_image)

        if self.transform is not None:
          input_image = self.transform(input_image)
          target_image = self.transform(target_image)

        #display(input_image)
        input_image = np.array(input_image)
        target_image = np.array(target_image)

        return torch.Tensor(input_image), torch.Tensor(target_image) 
    
    def __len__(self):
        return self.len
```
