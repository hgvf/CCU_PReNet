```python
import os
import numpy as np
import random
import torch
import cv2
import torch.utils.data as udata
import torch.utils.data.dataset as Datasets
from PIL import Image
from torchvision import transforms
from IPython.display import display
from matplotlib import pyplot as plt
```

```python
def normalize(data):
  return data/255.0
```

```python
# 自訂 datasets

class PReNet_train_datasets(torch.utils.data.Dataset):
    
    def __init__(self, transform=None):
               
        # datasets 存放路徑
        self.path = ''
        self.input_path = os.path.join(self.path, 'RAIN')
        self.target_path = os.path.join(self.path, 'GT')
        
        imgs = []

        # 蒐集所有檔名
        files = os.listdir(self.input_path)
        
        for f in files:
            imgs.append((f, int(0)))
        
        self.imgs = imgs
        self.len = len(imgs)
        self.transform = transform
        self.files = files

    def __getitem__(self, index):
        
        path, label = self.imgs[index]
        
        # File open
        # input image
        input_path = os.path.join(self.input_path, self.files[index])
        input_image = cv2.imread(input_path)
      
        # target image
        target_path = os.path.join(self.target_path, self.files[index])
        target_image = cv2.imread(target_path)

        # input image preprocessing
        b, g, r = cv2.split(input_image)
        input_image = cv2.merge([r, g, b])
        input_image = np.float32(normalize(input_image))
        input_image = input_image.transpose(2,0,1)
        
        # target image preprocessing
        b, g, r = cv2.split(target_image)
        target_image = cv2.merge([r, g, b])
        target_image = np.float32(normalize(target_image))
        target_image = target_image.transpose(2,0,1)

        # Transform
        if self.transform is not None:
          input_image = self.transform(input_image)
          target_image = self.transform(target_image)

        return (input_image, target_image)
    
    def __len__(self):
        return self.len
```
